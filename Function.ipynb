## Map, Reduce, Accumalate, lambda :

```python
a = ["1","2","3"]
b = list(map(float,a))
print(b)
```

```python
l = [2,3,5,4]
newl = list(map(cube,l))
print(newl)
```

```python
l = [2,3,5,4]
newl = list(map(lambda a: a+2,l))
print(newl)
```

```python
l = [2,3,5,4]
newl = list(filter(lambda a: a>3,l))
print(newl)
```

```python
from functools import reduce
def add(x,y):
    return x-y
    
l = [2,3,5,4]
res = reduce(add,l)
print(res)
```

```python
l = [2,3,5,4]
newl = list(map(lambda a: a+2,l))
print(newl)
```

```python
l = [2,3,5,4]
newl = list(map(lambda a:a**2,l))
print(newl)
```

```python
from functools import reduce
l = [2,3,5,4]
newl = list(reduce(lambda x,y:x+y,l))
print(newl)
```
